```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x=np.arange(-10,11)
```

```python
plt.figure(figsize=(12, 6))

plt.title('My Nice Plot')

plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10,0,10],[0,0,0])
plt.plot([0,0,0],[-100,0,100])
plt.legend(['X^2','-X^2','Horizontal line','Vertical line'])
plt.xlabel('X')
plt.ylabel('X squared')
```

```python
f=plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

plt.subplot(1, 2, 1,facecolor='lime',projection='lambert')  # rows, columns, panel selected
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2,facecolor='yellow',projection='polar')
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])

plt.xlabel('X')
plt.ylabel('X Squared')
```

# OOP interface

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(x, (x ** 2), color='red', linewidth=3,marker='*', markersize=9, label='X^2')

axes.plot(x, -1 * (x ** 2), 'b--',linewidth=2, label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title("My Nice Plot")

axes.legend()
fig
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid',label="solid")
axes.plot(x, x + 1, linestyle='dashed',label="dashed")
axes.plot(x, x + 2, linestyle='dashdot',label="dashdot")
axes.plot(x, x + 3, linestyle='dotted',label="dotted")
axes.plot(x, x/2,   linestyle='-',label="dashed");

axes.set_title("My Nice Plot")
axes.legend()
```

```python
fig, axes=plt.subplots(figsize=(12,6))
axes.plot(x, x+0, '-og',label="solid green")
axes.plot(x, x+1, '-.b', label="dashedotblue")
axes.plot(x, x+2, ':r' ,marker='v',markersize=10 , label="dotted red")
axes.plot(x, x+3, '--c',marker='D',markersize=10 , label="dashed cyan")
axes.plot(x, x+4, '--m',marker='>',markersize=10 , label="dashed purple");

axes.set_title("MY LITTLE NICE PLOT")

axes.legend()
```

```python
"""We can also define how many elements we want inside our figure. To do that we can set the nrows and ncols params."""
plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(14, 6))

fig, ((ax1, ax2), (ax3,ax4)) = plot_objects

plot_objects
```

```python
ax4.plot(np.random.randn(50), c='yellow')
ax1.plot(np.random.randn(50), c='red', linestyle='--')
ax2.plot(np.random.randn(50), c='green', linestyle=':')
ax3.plot(np.random.randn(50), c='blue', marker='<', linewidth=3.0)


fig
```
